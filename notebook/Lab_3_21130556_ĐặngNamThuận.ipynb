image.png

`y_pred = regressor.predict(X_test)`

*   metrics.explained_variance_score(y_test, y_pred)
*   metrics.mean_absolute_error(y_ test, y_pred)
*   metrics.mean_squared_error(y_ test, y_pred[, ...])
*   metrics.mean_squared_log_error(y_ test, y_pred)
*   metrics.median_absolute_error(y_ test, y_pred)

# Task 2.1. dataset1.csv


```python
dataset1 = pd.read_csv("dataset1.csv")
regressor = LinearRegression()
regressor.fit(dataset1[['x']], dataset1[['y']])
y_pred = regressor.predict(dataset1[['x']])
#regressor.intercept_
#regressor.coef_
#y_pred - dataset1[['y']]
#metrics.r2_score(dataset1[['y']], y_pred)

plt.scatter(dataset1[['x']], dataset1[['y']], s=30, color="blue", label="Real value")


plt.plot(dataset1[['x']], y_pred, color="red", label="Predict value")

plt.xlabel('x - axis')

plt.ylabel('y - axis')

plt.title('Predict graph for dataset1.csv!')

plt.legend()

plt.show()
```

# Task 2.2. dataset2.csv


```python
dataset2 = pd.read_csv("dataset2.csv")
regressor = LinearRegression()
X_train, X_test, y_train, y_test = train_test_split(dataset2[['x1', 'x2', 'x3']], dataset2[['y']], test_size=0.3, train_size=0.7, random_state=None, shuffle=True, stratify=None)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
print('Wo: ', regressor.intercept_)
print('W1, W2, W3: ', regressor.coef_)
print('explained_variance_score: ', metrics.explained_variance_score(y_test, y_pred))
print('mean_absolute_error: ', metrics.mean_absolute_error(y_test, y_pred))
print('mean_squared_error: ', metrics.mean_squared_error(y_test, y_pred))
#print('mean_squared_log_error: ', metrics.mean_squared_log_error(y_test, y_pred))
print('median_absolute_error: ', metrics.median_absolute_error(y_test, y_pred))
print('r2_score: ', metrics.r2_score(y_test, y_pred))
```

*   **plot y_pred** and **y_test** to show the different between these values

# Task 2.3. Apply LinearRegression to the following dataset:
*   Red wine quality: https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv . The dataset includes info about the chemical properties of different types of wine and how they relate to overall quality.



```python
winedata = pd.read_csv("winequality-red.csv")
regressor = LinearRegression()
X_train, X_test, y_train, y_test = train_test_split(winedata[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']], winedata[['quality']], test_size=0.3, train_size=0.7, random_state=None, shuffle=True, stratify=None)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
print('Wo: ', regressor.intercept_)
print('W1, W2, W3,..., W10, W11: ', regressor.coef_)
print('explained_variance_score: ', metrics.explained_variance_score(y_test, y_pred))
print('mean_absolute_error: ', metrics.mean_absolute_error(y_test, y_pred))
print('mean_squared_error: ', metrics.mean_squared_error(y_test, y_pred))
print('mean_squared_log_error: ', metrics.mean_squared_log_error(y_test, y_pred))
print('median_absolute_error: ', metrics.median_absolute_error(y_test, y_pred))
print('r2_score: ', metrics.r2_score(y_test, y_pred))
```

*   **plot y_pred** and **y_test** to show the different between these values

# Task 2.4. Apply LinearRegression to the following dataset:
*   Energy efficiency: https://archive.ics.uci.edu/ml/datasets/Energy+efficiency.




```python
dataset3 = pd.read_csv("ENB2012_data.csv")
regressor = LinearRegression()
X_train, X_test, y_train, y_test = train_test_split(dataset3[['X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X7', 'X8']], dataset3[['Y1', 'Y2']], test_size=0.3, train_size=0.7, random_state=None, shuffle=True, stratify=None)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
print('WoY1: ', regressor.intercept_[0])
print('WoY2: ', regressor.intercept_[1])
print('W1, W2, W3,..., W7, W8 of Y1: ', regressor.coef_[0])
print('W1, W2, W3,..., W7, W8 of Y2: ', regressor.coef_[1])
print('explained_variance_score: ', metrics.explained_variance_score(y_test, y_pred))
print('mean_absolute_error: ', metrics.mean_absolute_error(y_test, y_pred))
print('mean_squared_error: ', metrics.mean_squared_error(y_test, y_pred))
print('mean_squared_log_error: ', metrics.mean_squared_log_error(y_test, y_pred))
print('median_absolute_error: ', metrics.median_absolute_error(y_test, y_pred))
print('r2_score: ', metrics.r2_score(y_test, y_pred))
```

*   **plot y_pred** and **y_test** to show the different between these values
